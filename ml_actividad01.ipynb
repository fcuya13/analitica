6. Agrega una nueva columna 'precio_por_area' que calcule el precio por metro cuadrado (precio / area). (1 pts)

```python
#Quitar prefijos y sufijos para convertirlos a float
data['precio_float']=data['precio'].str.replace('S/ ','').str.replace('$ ','').str.replace(',','').astype(float)
data['area_float']=data['area'].str.replace(' m2','').astype(float)
#Crear nueva columna
data['precio_por_area']= data['precio_float']/data['area_float']
#Mostrar
data
```

7. Agrega una columna `moneda` donde se indique la moneda en la que se encuentra el precio. (0.5pts)

```python
data['moneda'] = data['precio'].str.extract(r'(S/|\$)')[0].apply(lambda x: 'PEN' if 'S/' in x else 'USD')
data
```

8. Agrupa el dataset por `distrito` y calcula el precio promedio por distrito. (1 pts)

```python
#Crear columna para uniformizar el precio por moneda
def convertirasoles(row):
    if row['moneda']=='USD':
        return row['precio_float']*3.7
    else:
        return row['precio_float']

data['precio_soles']=data.apply(convertirasoles, axis=1)

#Agrupar por distrito y calcular el promeido
data.groupby('distrito')['precio_soles'].mean()
```

9. ¿Cuál es precio promedio de los departamentos, por distrito, de tipo `flat` y que tienen `gimnasio` como `areas_comunes`? La tabla de salida debe estar ordenada alfabéticamente por el nombre del distrito. (1 pts)

```python
data[(data['tipo']=='flat')&(data['areas_comunes'].apply(lambda x: 'Gimnasio' in x))].groupby('distrito')['precio_soles'].mean().sort_index()
```

10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
X=data['area_float']
Y=data['precio_soles']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.1,random_state=27)

X_train_reshaped= X_train.values.reshape(-1,1)
Y_train_reshaped= Y_train.values.reshape(-1,1)
X_test_reshaped=X_test.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train_reshaped,Y_train_reshaped)
Y_pred=model.predict(X_test_reshaped)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
plt.figure(figsize=(10,6))
plt.scatter(data['area_float'], data['precio_soles'], color='pink')
plt.scatter(X_test, Y_test, color='purple')
plt.scatter(X_test,Y_pred, color='red')
plt.plot(X_train, model.predict(X_train_reshaped), color='black')
plt.gca().yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.0f}'))
plt.xlabel('Area')
plt.ylabel('Precio')
```

```python
w= model.coef_[0][0]
b= model.intercept_[0]
print("w= ", w)
print("b= ",b)
```

11.  ¿Cómo varía los coeficientes `w` y `b` dependiendo del distrito? Seleccione 5 casos y analice `w` y `b`. (2 pts)

```python
distritos=['Ate', 'Barranco', 'Bellavista', 'Breña', 'Cayma']
for distrito in distritos:
    datadistrito= data[data['distrito']==distrito]
    X=datadistrito['area_float']
    Y=datadistrito['precio_soles']

    X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.1,random_state=2)
    X_train_reshaped= X_train.values.reshape(-1,1)
    Y_train_reshaped= Y_train.values.reshape(-1,1)
    X_test_reshaped=X_test.values.reshape(-1,1)

    model=LinearRegression()
    model.fit(X_train_reshaped,Y_train_reshaped)
    Y_pred=model.predict(X_test_reshaped)

    print("w del distrito", distrito, ":", model.coef_[0][0])
    print("b del distrito", distrito, ":", model.intercept_[0],"\n")
```

### Crítica
Desde hace algunos años, la inteligencia artificial comenzó a implementarse en el sector de empleabilidad en el Perú, y aunque se han logrado avances importantes, todavía se encuentra en una etapa inicial. Según un estudio de la plataforma de empleo Bumeran, Perú es el país de latinoamérica que más utiliza inteligencia artificial en la gestión de Recursos Humanos (Forbes, 2024). Sin embargo, el uso de esta tecnología se limita principalmente a tareas operativas (como la búsqueda de candidatos en bases de datos y la creación automatizada de avisos de empleo) y aún no se aprovecha todo su potencial desde una visión estratégica en el sector. En el sector público, la bolsa de trabajo lanzada por el gobierno peruano es una iniciativa concreta que implementa IA para conectar a candidatos con empleadores al encontrar coincidencias entre los requisitos de puestos de trabajo y la información personal de los candidatos, como sus estudios, habilidades y experiencia (Ministerio de Trabajo y Promoción del Empleo, 2022). Aunque esta iniciativa representa un avance importante en innovación, no hay información sobre cómo se ha integrado esta tecnología con la plataforma que ya existía previamente o las técnicas usadas, lo cual hace difícil evaluar su calidad. Además, no se plantean métricas que evalúen el impacto de la IA. Por ejemplo, aunque se menciona el aumento del 24% en vacantes respecto al año anterior, no se especifica cómo la IA ha contribuido directamente a este aumento.

Por otro lado, hay diversas propuestas de proyectos que usan IA aplicables al sector empleo en el Perú. Los proyectos de selección de personal docente en una universidad nacional peruana (Yaranga, 2021) y de selección de personal administrativo en la Corte Superior de Justicia  de Lima (Coronel, 2020) tienen como objetivo común mejorar la eficiencia en los procesos de contratación. Ambos tienen el potencial de automatizar procesos y reducir el sesgo humano; no obstante, si no se entrenan los modelos adecuadamente, podrían estar sesgados por los datos usados y caer en discriminación. También, se ha planteado un proyecto para el análisis de la demanda en el sector TI (Mamani, 2022), lo cual favorece el análisis de datos para la toma de decisiones. Como todo modelo, estas tres iniciativas tienen el riesgo de hacer predicciones erróneas e imprecisas. En cuanto a Tinkiy (Peñafiel et al., 2024), busca mejorar la calidad de las decisiones relacionadas con la selección de personal y la asignación de oportunidades laborales. Sin embargo, este proyecto usa la IA de forma limitada, por ejemplo, para la generación de CVs, descripción de perfiles laborales o filtros de búsqueda. Por último, a pesar de que se están haciendo esfuerzos para mejorar el acceso a datos abiertos, como se muestra en la Plataforma Nacional de Datos Abiertos, aún hay limitaciones de la cantidad y calidad de data disponible, por lo que es importante considerar estas restricciones al considerar los posibles resultados de todos estos proyectos.
